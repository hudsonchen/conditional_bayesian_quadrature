```python
import numpy as np
from scipy.stats import norm
from sobol_seq import i4_sobol_generate
import matplotlib.pyplot as plt

```

```python
n = 200
u1 = i4_sobol_generate(2, n)
u2 = np.random.uniform(size=(n,2))

plt.figure()
plt.scatter(u1[:,0], u1[:,1], color='r')
plt.scatter(u2[:,0], u2[:,1], color='b')
plt.show()

mu = np.array([0.3, 0.5])
sigma = np.array([[1.0, 0.5],
                  [0.5, 1.0]])

L = np.linalg.cholesky(sigma)

x1 = mu[:, None] + (norm.ppf(u1) @ L).T
x2 = mu[:, None] + (norm.ppf(u2) @ L).T

plt.figure()
plt.scatter(x1[0, :], x1[1, :], color='r')
plt.scatter(x2[0, :], x2[1, :], color='b')

plt.show()
```
