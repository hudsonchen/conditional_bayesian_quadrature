```python
x0 = 0
mu_y_x = 0.8 * x1
cov_y_x = 1 - 0.8 * 0.8
precision_y_x = cov_y_x ** (-1)
eps = 1e-10

from sklearn.gaussian_process.kernels import RBF

MC_list = []
BMC_list = []

def int_K(mu, cov, y, sigma):
    return (1 + cov * sigma ** (-2)) ** (-0.5) * np.exp(-0.5 * (sigma ** 2 + cov) ** (-1) * (y - mu) ** 2)
    
N = 100
sigma = 0.1

for n in range(1, N+1):
    y_samples = np.random.normal(loc=mu_y_x, scale=np.sqrt(cov_y_x), size=n)[:, None]
    gY = g(y_samples)
    int_MC = gY.mean()
    
    K_mean = int_K(mu_y_x, cov_y_x, y_samples, sigma)
    K = RBF(length_scale = sigma)(y_samples, y_samples) + eps * np.eye(n)
    gY = g(y_samples)
    int_BMC = (K_mean.T @ np.linalg.inv(K) @ gY)[0]
    
    MC_list.append(int_MC)
    BMC_list.append(int_BMC)
    
plt.figure()
plt.plot(np.arange(1, 101), MC_list, color = 'b', label='MC')
plt.plot(np.arange(1, 101), BMC_list, color = 'r', label='BMC')
plt.legend()
plt.show()

print("This is the sanity check of using BMC for conditional expectation.")
```

```python
N = 100
sigma = 0.1

Nx = 4
x_obs = np.random.uniform(low=0.0, high=1.0, size=Nx)[:, None]
# x_obs = np.array([0.1, 0.2, 0.3, 0.4])[:, None]
x_pred = 0.5
eps = 1e-10

MC_list = []
mu_y_x = 0.8 * x_pred
cov_y_x = 1 - 0.8 * 0.8
precision_y_x = cov_y_x ** (-1)
# The MC is based on the conditioning distribution on the true value.
for n in range(1, N+1):
    y_samples = np.random.normal(loc=mu_y_x, scale=np.sqrt(cov_y_x), size=n)[:, None]
    gY = g(y_samples)
    int_MC = gY.mean()
    MC_list.append(int_MC)
    
BMC_list = []
for n in range(1, N+1):
    int_BMC_obs = []
    for x in x_obs:
        mu_y_x = 0.8 * x
        cov_y_x = 1 - 0.8 * 0.8
        precision_y_x = cov_y_x ** (-1)
        y_samples = np.random.normal(loc=mu_y_x, scale=np.sqrt(cov_y_x), size=n)[:, None]
    
        K_mean = int_K(mu_y_x, cov_y_x, y_samples, sigma)
        K = RBF(length_scale = sigma)(y_samples, y_samples) + eps * np.eye(n)
        gY = g(y_samples)
        int_BMC_obs.append((K_mean.T @ np.linalg.inv(K) @ gY)[0])
    
    int_BMC_obs = np.array(int_BMC_obs)
    x_pred = np.array([0.5])[:, None]
    Kx = RBF(length_scale = 0.5)(x_obs, x_obs) + eps * np.eye(len(x_obs))
    int_BMC = (RBF(length_scale = 0.5)(x_pred, x_obs) @ np.linalg.inv(Kx) @ int_BMC_obs)[0]
    BMC_list.append(int_BMC)

plt.figure()
plt.plot(np.arange(2, N+1), MC_list[1:], color = 'b', label='MC')
plt.plot(np.arange(2, N+1), BMC_list[1:], color = 'r', label='BMC')
plt.legend()
plt.show()

print(x_obs)
```

```python
import torch
from torch.autograd import grad
from torch import optim


def RBF_kernel(x_tensor, y_tensor, sigma):
    r = torch.cdist(x_tensor, y_tensor, p=2)
    return torch.exp(- 0.5 * (sigma ** (-2)) * (r ** 1))

n_max = 50
l_list = []

eps_1 = 1e-10

for n in tqdm(range(1, n_max+1)):
    int_BMC_obs = []
    y_samples_all = torch.zeros([n * len(x_obs), 1])
    
    for i, x in enumerate(x_obs):
        mu_y_x = 0.8 * x
        cov_y_x = 1 - 0.8 * 0.8
        precision_y_x = cov_y_x ** (-1)
        y_samples = torch.normal(mean=mu_y_x[0], std=np.sqrt(cov_y_x), size=(n,1))
        y_samples_all[i * n:(i+1) * n, :] = y_samples
        
    gy = g_torch(y_samples)

    log_l = torch.tensor([-1.0], requires_grad=True)
    optimizer = optim.Adam([log_l], lr=0.01)
    llk_curr = torch.tensor([0.0])
    llk = torch.tensor([1.0])

    while (torch.abs(llk_curr - llk) > 0.001):
        llk_curr = llk
        l = torch.exp(log_l)
        K = RBF_kernel(y_samples, y_samples, l)
        K_inv = torch.inverse((K + 0.01 * torch.eye(n * 1)))
        llk = (-0.5 * gy.T @ K_inv @ gy - 0.5 * torch.log(torch.det(K) + eps)) / (n * 1)
        nll = -llk
        nll.backward()
        optimizer.step()
        # print(llk.detach())
        # print(torch.exp(log_l))
    l_list.append(torch.exp(log_l).detach().numpy())

plt.figure()
plt.plot(range(1, n_max+1), l_list)
plt.title("The optimization of length scale under maximum log-likelihood.")
plt.xlabel("The number of observations.")
plt.ylabel("length scale")
plt.show()
```

```python
N = 100
sigma = 0.1

Nx = 4
x_obs = np.random.uniform(low=0.0, high=1.0, size=Nx)[:, None]
# x_obs = np.array([0.1, 0.2, 0.3, 0.4])[:, None]
x_pred = 0.5
eps = 1e-10
eps_1 = 1e-6

MC_list = []
mu_y_x = 0.8 * x_pred
cov_y_x = 1 - 0.8 * 0.8
precision_y_x = cov_y_x ** (-1)
# The MC is based on the conditioning distribution on the true value.
for n in range(1, N+1):
    y_samples = np.random.normal(loc=mu_y_x, scale=np.sqrt(cov_y_x), size=n)[:, None]
    gY = g(y_samples)
    int_MC = gY.mean()
    MC_list.append(int_MC)
    
BMC_list = []
l_list = []
for n in tqdm(range(1, N+1)):
    int_BMC_obs = []

    for i, x in enumerate(x_obs):
        mu_y_x = 0.8 * x
        cov_y_x = 1 - 0.8 * 0.8
        precision_y_x = cov_y_x ** (-1)
        y_samples_torch = torch.normal(mean=mu_y_x[0], std=np.sqrt(cov_y_x), size=(n,1))
        
    gy = g_torch(y_samples_torch)

    log_l = torch.tensor([1.0], requires_grad=True)
    optimizer = optim.Adam([log_l], lr=0.01)
    llk_curr = torch.tensor([0.0])
    llk = torch.tensor([1.0])

    while (torch.abs(llk_curr - llk) > 0.001):
        llk_curr = llk
        l = torch.exp(log_l)
        K = RBF_kernel(y_samples_torch, y_samples_torch, l)
        K_inv = torch.inverse((K + eps_1 * torch.eye(n * 1)))
        llk = (-0.5 * gy.T @ K_inv @ gy - 0.5 * torch.log(torch.det(K) + eps)) / (n * 1)
        nll = -llk
        nll.backward()
        optimizer.step()

    sigma = min(torch.exp(log_l).detach().numpy(), 0.1)

    for x in x_obs:
        mu_y_x = 0.8 * x
        cov_y_x = 1 - 0.8 * 0.8
        precision_y_x = cov_y_x ** (-1)
        y_samples = np.random.normal(loc=mu_y_x, scale=np.sqrt(cov_y_x), size=n)[:, None]
    
        K_mean = int_K(mu_y_x, cov_y_x, y_samples, sigma)
        K = RBF(length_scale = sigma)(y_samples, y_samples) + eps * np.eye(n)
        gY = g(y_samples)
        int_BMC_obs.append((K_mean.T @ np.linalg.inv(K) @ gY)[0])
    
    int_BMC_obs = np.array(int_BMC_obs)
    x_pred = np.array([0.5])[:, None]
    Kx = RBF(length_scale = 0.5)(x_obs, x_obs) + eps * np.eye(len(x_obs))
    int_BMC = (RBF(length_scale = 0.5)(x_pred, x_obs) @ np.linalg.inv(Kx) @ int_BMC_obs)[0]
    BMC_list.append(int_BMC)
    l_list.append(sigma)

plt.figure()
plt.plot(np.arange(2, N+1), MC_list[1:], color = 'b', label='MC')
plt.plot(np.arange(2, N+1), BMC_list[1:], color = 'r', label='BMC')
plt.legend()
plt.show()

print("The location of observed x", x_obs)
print("The location of x' is 0.5")
```
